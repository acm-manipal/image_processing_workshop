```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 

frame = cv2.imread("D:\lena_1.jpg")#sample image here


    # Convert BGR to HSV
frame = cv2.cvtColor(np.array(frame), cv2.COLOR_BGR2RGB)    
hsv = cv2.cvtColor(np.array(frame), cv2.COLOR_RGB2HSV)


plt.subplot(121)
plt.imshow(frame)

plt.subplot(122)
plt.imshow(hsv)

plt.show()
#cv2.destroyAllWindows()
```
