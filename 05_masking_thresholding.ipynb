```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images/image1.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

Loading our image and converting it to `HSV`

We use `HSV` for finding colour since `Hue` is a better measure of colour than what we can get using `RGB`

Let's say we want to find all green objects our image. We first find the required `HSV` value for green

```python
green = np.uint8([[[0,255,0 ]]])
hsv_green = cv2.cvtColor(green,cv2.COLOR_BGR2HSV)
print(hsv_green)
```

Now you take `[H-10, 100, 100]` and `[H+10, 255, 255]` as lower bound and upper bound respectively

```python
lower_green = np.array([50, 100, 100])
upper_green = np.array([70, 255, 255])
```

```python
green_mask = cv2.inRange(hsv, lower_green, upper_green)
```

`inRange` removes all values which are not in the interval between `lower_green` and `upper_green`

```python
res = cv2.bitwise_and(img, img, mask = green_mask)
```

We overlay `green_mask` over the original image `img` to get only the green parts

```python
plt.imshow(green_mask)
plt.show()
```

This is the mask that was extracted in `HSV` colourspace

```python
plt.imshow(res)
plt.show()
```
