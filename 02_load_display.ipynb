# Moving on...
let's figure out the basics -- loading and displaying an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

We import the required packages, matplotlib is used for displaying the image in a Jupyter notebook because OpenCV doesn't work :(

```python
img = cv2.imread("images/image1.jpeg")
```

We load the image "image1.jpeg" from /images and store it in a NumPy array

```python
plt.imshow(img)
plt.show()
```

We disply the image, but it doesn't look right.

The colours of our image are clearly wrong, this is because of a caveat with OpenCV.

OpenCV represents RGB images as multi-dimensional NumPy array but in reverse order. This means that the images are actually BGR order rather than RGB. 

There's an easy fix though.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

We can use cvtColor to convert from one colour space to the other. In this case, we use it to go from BGR to RGB

```python
plt.imshow(img_rgb)
plt.show()
```

Fixed!
